```python
import pandas as pd
import numpy as np
```

```python
key_data = pd.read_csv('Key_Data.csv')
```

```python
mentee_birth = key_data[['mentee_id','birthYear_mte']].drop_duplicates().rename(columns={'mentee_id':'author_id'})
mentee_birth.head(5)
```

```python
au_pa_year = pd.read_csv(f'author_paper_year.csv')
```

```python
au_pa_year.head(5)
```

```python
mte_mto = key_data[['mentee_id','mentor_id']].drop_duplicates().values.tolist()
print(len(mte_mto))
```

```python
len(mte_mto)
```

```python
Data = []
n0 = 0
n1 = n0 + 5000
for i in range(n0,n1):
    mte_i = mte_mto[i][0]
    mto_i = mte_mto[i][1]
    
    mte_i_pa = set(au_pa_year['work_id'][au_pa_year.author_id==mte_i])
    mto_i_pa = set(au_pa_year['work_id'][au_pa_year.author_id==mto_i])
    colla_work = mte_i_pa.intersection(mto_i_pa)
    colla_work_year = au_pa_year[['work_id','publication_year']][au_pa_year.work_id.isin(colla_work)].drop_duplicates()
    mte_i_birth = mentee_birth['birthYear_mte'][mentee_birth.author_id==mte_i].values[0]
    colla_work_year['career_stage'] = colla_work_year['publication_year'] - mte_i_birth
    colla_work_first5 = colla_work_year[colla_work_year.career_stage<=5].drop_duplicates()
    
    mte_i_pa_year = au_pa_year[['work_id','publication_year']][au_pa_year.author_id==mte_i].drop_duplicates()
    mte_i_pa_year['career_stage'] = mte_i_pa_year['publication_year'] - mte_i_birth
    mte_work_first5 = mte_i_pa_year[mte_i_pa_year.career_stage<=5]

    colla_work_count = len(colla_work_year)                              # 1
    colla_work_count_first5 = len(colla_work_first5)                     # 2
    colla_work_count_later5 = colla_work_count - colla_work_count_first5 # 3


    collaborators_mte = set(au_pa_year['author_id'][au_pa_year.work_id.isin(mte_i_pa)])
    collaborators_mto = set(au_pa_year['author_id'][au_pa_year.work_id.isin(mto_i_pa)])
    common_cllaborators = collaborators_mte.intersection(collaborators_mto)

    collaborators_mte_count = len(collaborators_mte)                              # 4
    common_cllaborators_count = len(common_cllaborators)                          # 5
    ratio_common_collaborators = len(common_cllaborators)/len(collaborators_mte)  # 6
    
    mte_work_first5_count = len(mte_work_first5)   # 7
    
    Data.append([mte_i,mto_i,colla_work_count,colla_work_count_first5,colla_work_count_later5,
                collaborators_mte_count,common_cllaborators_count,ratio_common_collaborators,
                mte_work_first5_count])
    
    if i%10==0:
        print(i)
```

```python
Data = pd.DataFrame(Data,columns=['mentee_id','mentor_id','colla_work_count','colla_work_count_first5','colla_work_count_later5',
                'collaborators_mte_count','common_cllaborators_count','ratio_common_collaborators','mte_work_first5_count'])
```
